```python
import pandas as pd
import matplotlib.pyplot as plt

# Count churn by SeniorCitizen
count_df = (
    df.groupby(['SeniorCitizen', 'Churn'])
    .size()
    .unstack(fill_value=0)
)

# Convert to percentage
percent_df = count_df.div(count_df.sum(axis=1), axis=0) * 100
plt.figure(figsize=(2, 3.5))

bottom = None
for churn in percent_df.columns:
    bars = plt.bar(
        percent_df.index.astype(str),
        percent_df[churn],
        bottom=bottom,
        label=churn
       
    )

    # Add % labels inside bars
    for bar in bars:
        height = bar.get_height()
        if height > 3:  # avoid clutter
            plt.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f"{height:.1f}%",
                ha='center',
                va='center',
                fontsize=9,
                color='white'
            )

    bottom = percent_df[churn] if bottom is None else bottom + percent_df[churn]

plt.title("Churn by SeniorCitizen (%)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", bbox_to_anchor = (1.05,1))
plt.show()
```

```python
import pandas as pd 
df = pd.read_csv('Telco_Customer_Churn.csv')
df.columns.values
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
# df = pd.read_csv('your_file.csv')

# List of categorical columns (excluding 'customerID' and numeric columns)
categorical_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
                    'PhoneService', 'MultipleLines', 'InternetService',
                    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
                    'PaperlessBilling', 'PaymentMethod']

# Determine number of rows and columns for subplots
n_cols = 3
n_rows = (len(categorical_cols) + n_cols - 1) // n_cols

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(19, 5 * n_rows))
axes = axes.flatten()

# Plot countplots with bar labels using bar_label()
for i, col in enumerate(categorical_cols):
    ax = axes[i]
    bx = sns.countplot(data=df, x=col, hue='Churn', ax=ax)
    ax.set_title(col, fontsize=14)
    ax.tick_params(axis='x', rotation=45)
    
    # Add labels on bars
    for container in ax.containers:
        ax.bar_label(container, label_type='edge', fontsize=11)

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#Churn is not strongly influenced by gender, but is higher among senior citizens compared to non-senior customers.

#Customers without partners or dependents show noticeably higher churn rates than those with family ties.

#Month-to-month contracts, lack of online security/tech support, and fiber optic internet users are associated with higher churn.

#Customers using electronic check payment and paperless billing churn more, while long-term contracts and added services reduce churn.
